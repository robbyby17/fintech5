```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

#Impor dataset
df_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Fintech Demo Day/train.csv', sep=';')
df_test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Fintech Demo Day/test.csv', sep=';')
```

```python
# sample 10% of data to be used as unseen data
data = df_train.sample(frac=0.9, random_state=1708)
data_unseen = df_train.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
# print the revised shape
print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.classification import *
s = setup(data = data, target = 'default', session_id=123)
```

```python
best_model = compare_models()
```
